```python
string = input("Enter a string value: ")

char_count = {}

for ch in string:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1


for ch, count in char_count.items():
    print(f"{ch} = {count}", end=", ")
```

```python
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(find_max(34, 12, 7))
```

```python
def exponent(base, exp):
    return base ** exp

b = int(input("Enter the base: "))
e = int(input("Enter the exponent: "))

print(exponent(b, e))
```

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i**3
    return total

num = int(input("Enter a positive integer: "))
print(sum_of_cubes(num))
```

```python
for i in range(1, 10 + 1):
    if i % 2 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 2 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
numbers = [2, 3, 4, 2, 5, 2]

most_frequent = max(numbers, key=numbers.count)

print(most_frequent)
```

```python
numbers = [2, 1, 3, 1]

sum_squares = 0

for n in numbers:
    sum_squares += n * n

print(sum_squares)
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact = fact * i
    return fact

print(factorial(3))
```
