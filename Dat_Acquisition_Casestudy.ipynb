```python
import pandas as pd
import numpy as py
```

Step 1: Load SpaceX Launch Data from API

```python
launches_data=pd.read_json('/content/launches.json')
```

```python
launches_data.head()
```

```python
launches_data.info()
```

```python
#Extract relevant columns: name, date_utc, success, details, rocket
relevant_columns=launches_data[['name', 'date_utc', 'success', 'details', 'rocket']]
relevant_columns.head()
```

```python
#Convert date_utc to datetime and extract the year

relevant_columns['date_utc']=pd.to_datetime(launches_data['date_utc'])
```

```python
relevant_columns['year']=launches_data['date_utc'].astype(str).str.split('-').str[0]
```

```python
relevant_columns[['date_utc','year']].head()
```

Step 2: Load Rocket Metadata

```python
rockets_data=pd.read_json('/content/rockets.json')
```

```python
rockets_data.head()
```

```python
rockets_data.info()
```

```python
#Extract id, name, type, active, and stages
extract_columns=rockets_data[['id','name', 'type', 'active', 'stages']]
extract_columns.head()
```

Step 3: Merge Launch and Rocket Data

```python
#Join the two DataFrames on rocket ID using,Extract id, name, type, active, and stages
rockets_merge=extract_columns.set_index('id').add_suffix('_rocket')

merge_data=relevant_columns.join(rockets_merge,on='rocket')
merge_data
```

Step 4: Add Simulated Country Information

```python
#Add a new column country and randomly assign one of these values:['USA', 'Russia', 'India', 'China', 'France']
import numpy as np
merge_data['country'] = np.random.choice(['USA', 'Russia', 'India', 'China', 'France'], size=len(merge_data))
merge_data
```

Step 5: Store Merged Data in SQLite3

```python
#Use sqlite3 to create a connection and save the merged DataFrame as a table named launches
#Table should contain all merged columns including country
import sqlite3
```

```python
con=sqlite3.connect("spacex_launch_rocket.db")
```

```python
merge_data.to_sql("launches",con,index=False)
```

```python
launches_rockets_table=pd.read_sql("select * from launches",con)
launches_rockets_table
```

Step 6: Run SQL Queries on the Data to analyze

```python
# Launches by Country
cursor=con.cursor()
cursor.execute("""select country,count(*) as total_launches from launches group by country order by total_launches desc""")
```

```python
rows=(cursor.fetchall())
for row in rows:
  print(row)
```

```python
#  Which year had the highest number of launches?
import sqlite3

cursor.execute("""
    SELECT year, COUNT(*) AS launches_count
    FROM launches
    GROUP BY year
    ORDER BY launches_count DESC
    LIMIT 1;
""")
```

```python
row = cursor.fetchone()
```

```python
print(row)
```

```python
#  Top 5 Missions by Launch Count
cursor.execute("""
    SELECT name, COUNT(*) AS launch_count
    FROM launches
    GROUP BY name
    ORDER BY launch_count DESC
    LIMIT 5;
""")
```

```python
rows = cursor.fetchall()
for row in rows:
  print(row)
```

```python
#Top 3 by rocket count
cursor.execute("""
    SELECT name_rocket, COUNT(*) AS launch_count
    FROM launches
    GROUP BY name_rocket
    ORDER BY launch_count DESC
    LIMIT 5;
""")
```

```python
rows = cursor.fetchall()
for row in rows:
  print(row)
```
