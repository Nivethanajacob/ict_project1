```python
#sum of digits
num = input("Enter a number: ")

sum_digits = 0

for digit in num:
    sum_digits += int(digit)

print(f"Sum of digits:{sum_digits}")
```

```python
#factorial
def factorail(n):
  fact=1

  for i in range(1,n+1):
    fact *=i
  return fact

num1=int(input("Enter the number: "))
print(f"Factorial is:{factorail(num1)}")
```

```python
#palindrome
text=input("Enter a word: ")
cleaned = text.replace(" ", "").lower()

if cleaned==cleaned[::-1]:
  print("It is a palindrome")
else:
  print("It is not a palindrome")
```

```python
#vowels and consonants
text1=input("enter the word: ")
vowels="aeiouAEIOU"
v=0
c=0

for ch in text1:
  if ch in vowels:
    v +=1
  elif ch.isalpha():
    c +=1

print(f"vowels:{v}")
print(f"consonants:{c}")



```

```python
#common elements
list1=[1,2,3,4]
list2=[2,4,6,8]

common=[]

for x in list1:
  if x in list2:
    common.append(x)

print(f"common elements:{common}")
```

```python
#frequency
text2=input("Enter the string: ")

freq={}

for ch in text2:
  if ch in freq:
    freq [ch] +=1
  else:
    freq [ch] =1

print(freq)
```

```python
#leap year
year=int(input("Enter the year: "))

if year%4==0:
  print("It is a leap year")
else:
  print("It is not a leap year")
```

```python
#prime number
num = int(input("Enter a number: "))

if num < 2:
    print("Not a prime number")
else:
    prime = True

    for i in range(2, num):
        if num % i == 0:
            prime = False
            break

    if prime:
        print("Prime number")
    else:
        print("Not a prime number")
```

```python
# Fibonacci
n=int(input("Enter number of terms: "))

a,b=0,1

for i in range(n):
  print(a,end=" ")
  a,b=b,a+b
```

```python
#roots
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b*b - 4*a*c   # discriminant

if d >= 0:
    root1 = (-b + (d ** 0.5)) / (2*a)
    root2 = (-b - (d ** 0.5)) / (2*a)
    print("Roots are:", root1, "and", root2)
else:
    print("No real roots")
```
