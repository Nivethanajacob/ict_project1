# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def display_range(list_numbers):

  minimum_value=list_numbers[0]
  #setting the first value of the list as minimum value

  maximum_value=list_numbers[0]
   #setting the first value of the list as maximum value


  for number in list_numbers:
    #looping through the entire list to checkfor
    #maximum value and minimum value

    if(number>=maximum_value):
      maximum_value=number
    if(number<=minimum_value):
      minimum_value=number
  return(maximum_value,minimum_value)


my_list=[23,55,67,32,90]
range=display_range(my_list)
range

```

## Q2. Write function that returns mean value of a list of numbers

```python
def display_mean(list_number):
  sum=0

  for number in list_number:
    sum+=number
  mean=sum/len(list_number)
  return mean

my_list1=[1,2,3,4,5]
display=display_mean(my_list1)
display
```

# Q3. Write a function that returns the median value of a list of numbers

```python
def range_median(list_numbers):
    # sort the list
    sorted_list = sorted(list_numbers)

    length = len(sorted_list)

    # if the length is odd
    if length % 2 == 1:
        middle = sorted_list[length // 2]
        return middle

    # if the length is even
    else:
        middle1 = sorted_list[length // 2 - 1]
        middle2 = sorted_list[length // 2]
        return (middle1 + middle2) / 2

my_list2 = [1, 2, 3,6]
result = range_median(my_list2)
print(result)

```
